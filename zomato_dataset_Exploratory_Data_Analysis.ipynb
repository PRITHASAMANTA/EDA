```python
#Missing values
df.isnull().sum()
```

```python
#list comprehension : checking missing values
[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
# help of hitmap : missing values finding
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.info()
```

```python
df.columns
```

```python
# combine two country code
# ON : use for on which column u want to merge
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values= final_df.Country.value_counts().values
```

```python
# help of hitmap : missing values finding
country_values= final_df.Country.value_counts().values
## PIE CHART
plt.pie(country_values,labels=country_names,autopct='%1.2f%%')
plt.show()
```

```python
# pie chart of top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
# pie chart of least 3 countries
plt.pie(country_values[-3:],labels=country_names[-3:],autopct='%1.2f%%')
plt.show()
```

```python
# pie chart of  countries except first 3
plt.pie(country_values[3:],labels=country_names[3:],autopct='%1.2f%%')
plt.show()
```

```python
# pie chart of  countries from index 3 to 5
plt.pie(country_values[3:5],labels=country_names[3:5],autopct='%1.2f%%')
plt.show()
```

```python
# pie chart of countries from index 6
plt.pie(country_values[6:],labels=country_names[6:],autopct='%1.2f%%')
plt.show()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating)
```
